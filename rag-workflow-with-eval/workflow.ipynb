# RAG workflow
This notebook demonstrates a local Retrieval-Augmented Generation (RAG) workflow.
It orchestrates document ingestion, retrieval, and answer synthesis using LlamaIndex and Ollama,
allowing you to perform end-to-end question answering over your own documents with minimal setup.

## 1. Initialize the RAG Workflow

First, we create an instance of our `RAGWorkflow` class. All the complex setup of models and settings is handled internally.

```python
from workflow import RAGWorkflow

workflow = RAGWorkflow()

await workflow.ingest_documents("data")
```

## 2. Retrieve Relevant Documents

Next, we use the RAG workflow to retrieve the most relevant documents from the ingested index.
The workflow's retriever locates the top-matching source chunks based on your query, setting the stage for accurate answer generation.

```python
response = await workflow.query("How was DeepSeekR1 trained?")
print(response)
```
